```python
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
```

```python
data_train=pd.read_csv("/Users/alok.g.v/Desktop/python practice/Guvi Task/Task7/Assignment10_data/trainKNN.txt",sep=',',header=None,
                       names=['id number','refractive index','Na: Sodium','Mg: Magnesium','Al: Aluminum','Si: Silicon',' K: Potassium',
                             'Ca: Calcium','Ba: Barium','Fe: Iron','Type of glass'],index_col=0)
data_train
```

```python
x=['refractive index','Na: Sodium','Mg: Magnesium','Al: Aluminum','Si: Silicon',' K: Potassium',
                             'Ca: Calcium','Ba: Barium','Fe: Iron']
y=['Type of glass']
```

```python
X=data_train[x]
y=data_train[y]
```

```python
scaler=StandardScaler()
X=scaler.fit_transform(X)
print(X,y)
```

```python
data_test=pd.read_csv('/Users/alok.g.v/Desktop/python practice/Guvi Task/Task7/Assignment10_data/testKNN.txt',header=None,names=['id number','refractive index','Na: Sodium','Mg: Magnesium','Al: Aluminum','Si: Silicon',' K: Potassium',
                             'Ca: Calcium','Ba: Barium','Fe: Iron','Type of glass'],index_col=0)
data_test
```

```python
X_test=data_test[x]
y_test=data_test['Type of glass']
print(X_test,y_test)
```

```python
X_test=scaler.fit_transform(X_test)
X_test
```

```python
def knn_e(k):
    knc_e=KNeighborsClassifier(n_neighbors=k,metric='euclidean')
    knc_e.fit(X,y)
    y_pred=knc_e.predict(X_test)
    acc=accuracy_score(y_test,y_pred)
    print(f'Number of Neighbors : {k}  accuracy : {acc}')
        

    
```

```python
for x in range(1,11,2):
    knn_e(x)
```

```python
def knn_m(k):
    knc_m=KNeighborsClassifier(n_neighbors=k,metric='manhattan')
    knc_m.fit(X,y)
    y_pred=knc_m.predict(X_test)
    acc=accuracy_score(y_test,y_pred)
    print(f'Number of Neighbors : {k}  accuracy : {acc}')

```

```python
for x in range(1,11,2):
    knn_m(x)
```
